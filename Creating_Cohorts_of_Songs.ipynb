## Data Cleaning and Preprocessing

### Subtask:
Check for and handle any duplicate entries in the dataset, especially regarding song IDs, and remove the redundant `Unnamed: 0` column.

**Reasoning**:
As per the initial data inspection, there are no missing values. However, to ensure data quality and avoid biases in analysis, it's crucial to check for and remove any duplicate song entries. The `id` column is specified as a unique song identifier in the problem statement, making it the most appropriate column to check for duplicates. Additionally, the 'Unnamed: 0' column is a redundant index and should be dropped. I'll also convert `release_date` to datetime objects for proper temporal analysis and `duration_ms` to seconds for better interpretability.

```python
import pandas as pd

# Drop the 'Unnamed: 0' column as it is a redundant index
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
    print("Dropped 'Unnamed: 0' column.\n")

# Check for duplicate song IDs
duplicates_by_id = df.duplicated(subset=['id']).sum()
print(f"Number of duplicate song IDs found: {duplicates_by_id}")

if duplicates_by_id > 0:
    # Display some of the duplicate entries to understand their nature
    print("\nExamples of duplicate song IDs (first occurrence kept):\n")
    print(df[df.duplicated(subset=['id'], keep=False)].sort_values(by='id').head())

    # Remove duplicate entries based on 'id', keeping the first occurrence
    df_cleaned = df.drop_duplicates(subset=['id'], keep='first')
    print(f"\nRemoved {duplicates_by_id} duplicate entries based on 'id'.\n")
    print(f"New DataFrame shape after removing duplicates: {df_cleaned.shape}")
else:
    df_cleaned = df.copy()
    print("No duplicate song IDs found.\n")

# Convert 'release_date' to datetime objects
df_cleaned['release_date'] = pd.to_datetime(df_cleaned['release_date'])
print("Converted 'release_date' to datetime objects.\n")

# Convert 'duration_ms' to 'duration_s' (seconds)
df_cleaned['duration_s'] = df_cleaned['duration_ms'] / 1000
df_cleaned = df_cleaned.drop(columns=['duration_ms'])
print("Converted 'duration_ms' to 'duration_s' and dropped original column.\n")

# Display the info of the cleaned DataFrame to verify changes
print("### Info of the Cleaned DataFrame:\n")
df_cleaned.info()
print("\n" + "-"*50 + "\n")

# Display the first few rows of the cleaned DataFrame
print("### First 5 rows of the Cleaned DataFrame:\n")
print(df_cleaned.head())
```

### Subtask:
Consider the need for feature scaling or transformation if required for clustering.

**Reasoning**:
Clustering algorithms, especially distance-based ones like K-Means, are sensitive to the scale and distribution of features. Features with larger numerical ranges can dominate the distance calculations, leading to biased clustering results. Therefore, it's important to assess if our numerical features need scaling (e.g., Min-Max Scaling or Standardization) or transformation (e.g., log transformation for skewed data) to ensure all features contribute equally to the clustering process. Given the varying ranges observed in the univariate analysis (e.g., `loudness` being negative, `tempo` having a larger range compared to `acousticness`), scaling will likely be beneficial.

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

print("### Assessing the need for Feature Scaling\n")

# Select the numerical features that will be used for clustering and require scaling
# Exclude 'popularity' as it's our target for analysis, not necessarily a feature for clustering directly, unless specified.
# Exclude 'track_number', 'id', 'uri' as they are identifiers.
features_for_scaling = [
    'acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness',
    'loudness', 'speechiness', 'tempo', 'valence', 'duration_s'
]

# Display descriptive statistics before scaling to highlight the different scales
print("Descriptive statistics of features before scaling:\n")
print(df_cleaned[features_for_scaling].describe())
print("\n" + "-"*50 + "\n")

# Initialize the StandardScaler
scaler = StandardScaler()

# Create a copy of the DataFrame to store scaled features
df_scaled = df_cleaned.copy()

# Apply scaling to the selected features
df_scaled[features_for_scaling] = scaler.fit_transform(df_cleaned[features_for_scaling])

# Display descriptive statistics after scaling to show the effect
print("Descriptive statistics of features after StandardScaler scaling:\n")
print(df_scaled[features_for_scaling].describe())
print("\n" + "-"*50 + "\n")

print("Visualizing feature distributions after scaling (first few features):\n")
# Visualize the distributions after scaling to confirm (e.g., first 4 scaled features)
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_for_scaling[:4]): # Plot first 4 for illustration
    plt.subplot(2, 2, i + 1)
    sns.histplot(df_scaled[feature], kde=True, bins=30)
    plt.title(f'Scaled Distribution of {feature.replace("_", " ").title()}')
    plt.xlabel('Scaled Value')
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

print("Feature scaling completed. The features are now standardized, which will benefit clustering algorithms.")
```

Determine appropriate methods for identifying the optimal number of clusters (e.g., Elbow method, Silhouette score).

**Reasoning**:
Identifying the optimal number of clusters is a critical step in cluster analysis. Two common and effective methods are the Elbow Method and the Silhouette Score. The Elbow Method helps to find the point where adding more clusters does not significantly improve the explained variance, suggesting a natural number of clusters. The Silhouette Score measures how similar an object is to its own cluster (cohesion) compared to other clusters (separation), with higher values indicating better-defined clusters. I will use both methods to evaluate different numbers of clusters for our song data after applying PCA (using the components that explain 95% variance) to the scaled features.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

print("### Applying Principal Component Analysis (PCA)\n")

# Select the features for PCA (using the scaled features from df_scaled)
pca_features = features_for_scaling # Reusing the list of features that were scaled

X_scaled = df_scaled[pca_features]

# Initialize PCA
pca = PCA()

# Fit PCA on the scaled data
pca.fit(X_scaled)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Explained variance ratio: {explained_variance_ratio}\n")

# Cumulative explained variance
cum_explained_variance = np.cumsum(explained_variance_ratio)
print(f"Cumulative explained variance: {cum_explained_variance}\n")

# Plotting explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), cum_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

# Determine the number of components to retain (e.g., to explain 95% of variance)
num_components_95 = np.where(cum_explained_variance >= 0.95)[0][0] + 1
print(f"\nNumber of components to explain 95% of variance: {num_components_95}")

print("\nPCA analysis complete. The plot helps in deciding the optimal number of components for dimensionality reduction.")
```
